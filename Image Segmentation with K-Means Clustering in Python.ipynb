```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
image = mpl.image.imread('dog.jpg')
plt.imshow(image)
```

```python
image.shape
```

```python
x = image.reshape(-1, 3)
x.shape
```

```python
kmeans = KMeans(n_clusters=2, n_init=10)
kmeans.fit(x)
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
plt.imshow(segmented_img / 255)
```

```python
kmeans = KMeans(n_clusters=5, n_init=10)
kmeans.fit(x)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
plt.imshow(segmented_img / 255)
```

```python
kmeans = KMeans(n_clusters=50, n_init=10)
kmeans.fit(x)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
plt.imshow(segmented_img / 255)
```

```python
import cv2
cv2.imwrite('image1.png', cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
cv2.imwrite('image2.png', cv2.cvtColor(segmented_img.astype("uint8"), cv2.COLOR_BGR2RGB))
```
